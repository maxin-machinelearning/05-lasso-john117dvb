```python
from sklearn.linear_model import Lasso
import numpy as np 
import matplotlib.pyplot as plt
```

导入数据

```python
N = 100
X =np.linspace(-3,3,N).reshape(-1,1)
Y_real=2.4*X
Y=2.4*X+np.random.rand(N).reshape(-1,1)*4-2

#添加outliers
Y += 0.01 * np.random.normal()
```

sklearn实现LASSO

```python
clf = Lasso(alpha=1)
clf.fit(X,Y)
prediction = clf.predict(X)

print(clf.score(X,Y))
```

```python
plt.title("LASSO  alpha=1")
plt.plot(Y_real,color='b')
plt.plot(Y,color='yellow')
plt.plot(prediction,color='r')
plt.legend(['Y_real','Y','prediction'])
plt.show()
```

不同的正则化参数

```python
plt.subplot(1, 3, 1)
plt.title('alpha=2')
clf1 = Lasso(alpha=2)
clf1.fit(X,Y)
prediction1 = clf1.predict(X)
plt.plot(Y_real,color='b')
plt.plot(Y,color='yellow')
plt.plot(prediction1,color='r')
plt.legend(['Y_real','Y','prediction1'])


plt.subplot(1, 3, 2)
plt.title('alpha=0.5')
clf2 = Lasso(alpha=0.5)
clf2.fit(X,Y)
prediction2 = clf2.predict(X)
plt.plot(Y_real,color='b')
plt.plot(Y,color='yellow')
plt.plot(prediction2,color='r')
plt.legend(['Y_real','Y','prediction2'])


plt.subplot(1, 3, 3)
plt.title('alpha=0.001')
clf3 = Lasso(alpha=0.001)
clf3.fit(X,Y)
prediction3 = clf3.predict(X)
plt.plot(Y_real,color='b')
plt.plot(Y,color='yellow')
plt.plot(prediction3,color='r')
plt.legend(['Y_real','Y','prediction3'])
```

不同正则化参数下的拟合程度

```python
print(clf1.score(X,Y))
print(clf2.score(X,Y))
print(clf3.score(X,Y))
```

小结
1.能运用sklearn实现LASSO
2.添加数据异常值：在y中添加outliers
3.不同alpha下拟合程度不同
